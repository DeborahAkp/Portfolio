```python
def change_name(x): 
    if 'Director,' in x: 
        x = 'Director'
        return x
    elif 'Exec Assistant,'in x:
        x = 'Executive Assistant'
        return x
    elif x == 'CHief Information Officer':
        x = 'Chief Information Officer'
        return x
    elif 'VP'in x:
        x = 'VP'
        return x
    else: 
        x = x
        return x

df['job_mod'] = df['job_title'].apply(lambda x: change_name(x))
```

```python
df.groupby('job_mod')['job_mod'].count()
```
